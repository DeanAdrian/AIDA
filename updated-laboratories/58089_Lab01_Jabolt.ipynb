### Task 2: Getting Class Statistics
1. Create a function named `get_grades()` for computing the semestral grade of each student:

  `inputs`: `DataFrame` of a class grade sheet.
  
  `outpus`: `DataFrame` showing:
  * The prelim, midterms, and finals grades of each student
  * The semestral grade of each student



```python
def get_grades(dataFrame):
  ## Declaring of Variables
  grade_percent = np.array([0.3, 0.3, 0.4])
  grades_array = dataFrame[["prelim_grades", "midterm_grades", "final_grades"]].to_numpy()
  sem_grades = []

  ## For loop for Semestral Grades Calculation
  for grades in grades_array:
    grade_calc = np.dot(grade_percent, grades)
    sem_grades.append(grade_calc)

  ## Adding semestral_grades column to dataframe
  dataFrame['semestral_grades'] = sem_grades

  ## Return only specific columns after
  return dataFrame[['student_number', 'prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades']]
```

2. Create a function named `get_class_stats()`.
  
  `inputs`: `DataFrame` from `get_grades()`.
  
  `outpus`: `DataFrame` showing:
  * The lowest and highest prelim, midterm, finals, and semestral grades
  * the mean, median, mode, and standard deviation of the grades


```python
def get_class_stats(dataFrame):
  ##Declaring of Variables
  minValues = dataFrame[['prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades']].min()
  maxValues = dataFrame[['prelim_grades', 'midterm_grades', 'final_grades', 'semestral_grades']].max() 
  mean = dataFrame.mean()
  median = dataFrame.median()
  std = dataFrame.std()

  prelim_mode = dataFrame['prelim_grades'].mode()[0]
  midterm_mode = dataFrame['midterm_grades'].mode()[0]
  final_mode = dataFrame['final_grades'].mode()[0]
  semestral_mode = dataFrame['semestral_grades'].mode()[0]

  ##Creating new dataframe and assigning data
  new_df= pd.DataFrame(minValues)
  new_df.columns = ['lowest_grade']
  new_df['highest_grade'] = maxValues
  new_df['mean'] = mean
  new_df['median'] = median
  new_df['mode'] = [prelim_mode, midterm_mode, final_mode, semestral_mode]
  new_df['std_deviation'] = std
  
  ##Return new dataframe
  return new_df
```

```python
get_class_stats(df)
```
